```python
import time
from time import sleep
import requests
import random
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
import os
```

```python
import user_agents
from user_agents import DESKTOP_USER_AGENTS
```

```python
def send_request(link):
    time.sleep(random.choice(range(5,8)))
    driver.get(link)  

product_info_data = []
page = 1
url = 'https://www.ssense.com/en-us/women?q=bag'

#initiate webdriver
driver_path = "/Users/liutianyao/Downloads/chromedriver"
driver = webdriver.Chrome(driver_path)
res = send_request(url)

while page < 41:
    time.sleep(random.choice(range(3,10)))
    soup = BeautifulSoup(driver.page_source)
    product_infos = soup.findAll('figure',{'class':'browsing-product-item'})
    for p_info in product_infos:
        
        try: 
            brand = p_info.find('p',{'class':'bold'}).get_text()
        except: 
            brand = ""
            
        try: 
            product_name = p_info.find('p',{'class':'product-name-plp'}).get_text().replace('\n', '')
        except: 
            product_name = ""
            

        try:
            price =  p_info.find('p',{'class':'price'}).get_text().replace('$', '')
        except:
            price = ""

        new_entry = {
                    "Brand": brand,
                    "Product Name": product_name,
                    "Price": price
                    }
        product_info_data.append(new_entry)
        #use selenium driver to click next page
    try:
        button = driver.find_element_by_xpath('/html/body/div[1]/div/div/div[1]/main/div/div[1]/section/div[2]/div/div/nav/ul/li[8]')
        button.click()      
    except: 
        break 
    page += 1
    
driver.close()
```
